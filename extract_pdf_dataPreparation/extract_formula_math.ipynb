```python
from docxlatex import Document
import re, pandas as pd
import fitz
```

```python
def extract_fractions(text):
    # Regex pattern to match LaTeX \frac command
    frac_pattern = re.compile(r'\\frac{([^{}]+)}{([^{}]+)}')
    fractions = frac_pattern.findall(text)
    return fractions

def extract_formulas(file_path):
  doc = Document(file_path)
  text = doc.get_text()
  text = extract_fractions(text)
  text = text.split("\n")
  return {
      'text': text,
      'filename': [file_path for i in range(len(text))]
  }

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()

    text = text.split(" \n ")
    # text = [txt for txt in text if len(txt) > 0]
    return {
      'text': text,
      'filename': [pdf_path for i in range(len(text))]
  }
```

```python
file_path_original = 'drive/MyDrive/project-ugm/extract_pdf_dataPreparation/contoh_input/equation.pdf'

file_path = file_path_original.split('.')
file = None
if file_path[len(file_path)-1] == 'docx':
  file = extract_formulas(file_path_original)
elif file_path[len(file_path)-1] == 'pdf':
  file = extract_text_from_pdf(file_path_original)
  print(file)
```

```python
df = pd.DataFrame(file)
df.to_csv('drive/MyDrive/project-ugm/extract_pdf_dataPreparation/contoh_output_formula/output-docx.csv', index=False)
```
